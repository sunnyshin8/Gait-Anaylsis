```python
import pandas as pd
```

```python
df = pd.read_csv('synthetic_dataset.csv')
```

```python
df.info()
```

```python
#drop subject_id
df.drop('Subject_id', axis=1, inplace=True)
```

```python
#check correlation in heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')

#check correlation in table
df.corr()
```

```python
#train decision tree on all features target is the target row
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df.drop('Target', axis=1).values
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

#predict
predictions = dtree.predict(X_test)

#evaluate
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
#use all algorithms to find best one for this dataset (random forest, decision tree, naive bayes, svm, knn)
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
#random forest
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
print(accuracy_score(y_train, rfc.predict(X_train)))
print(accuracy_score(y_test, rfc.predict(X_test)))

#predict
rfc_pred = rfc.predict(X_test)

#evaluate
print(classification_report(y_test, rfc_pred))
xr = confusion_matrix(y_test, rfc_pred) 

#decision tree
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

#predict
dtree_pred = dtree.predict(X_test)

#evaluate
print(classification_report(y_test, dtree_pred))
xd = confusion_matrix(y_test, dtree_pred)

#naive bayes
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)


#predict
nb_pred = nb.predict(X_test)

#evaluate
print(classification_report(y_test, nb_pred))
xg = confusion_matrix(y_test, nb_pred)


```

```python
# Define a function to plot the confusion matrix
def plot_confusion_matrix(cm, labels, model_name):
    plt.figure(figsize=(4, 4))
    plt.title(model_name)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

# Replace labels_list with a list of your class labels (e.g., ['Class 0', 'Class 1'])
labels_list = ['0', '1', '2','3']  # Replace this with your actual class labels
plot_confusion_matrix(xg, labels_list, "GaussianNB")
```

```python
plot_confusion_matrix(xd, labels_list, "DecisionTreeClassifier")
```

```python
plot_confusion_matrix(xr, labels_list, "RandomForestClassifier")
```
